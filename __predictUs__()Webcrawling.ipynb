```python
from selenium import webdriver 
from selenium.webdriver.support.ui import Select 
from selenium.webdriver.common.by import By
import time 
import re 
import pandas as pd
```

```python
driver=webdriver.Chrome("X:/Python/chromedriver.exe")
    
driver.get("https://www.kickstarter.com/discover/advanced?sort=magic&seed=2555148&page=1")



elems = driver.find_elements_by_xpath(".//a[@class='soft-black hover-text-underline']")

allURLs = [] 


for elem in elems:
    allURLs.append(elem.get_attribute("href")) 

    
    
    
```

```python
goal=[]

for url in allURLs:
    
    driver.get(url)
    
    a= driver.find_elements_by_css_selector('span.money')[1].text
    
    goal.append(a)
                            
        
        
```

```python
deadline=[] 

for url in allURLs:
    driver.get(url)
    b = driver.find_elements_by_xpath("//p[contains(@class,'mb3')]")[0].text
    
    deadline.append(b)

print(deadline)
```

```python
deadline= [ (i.split(",")[1]) for i in deadline] 
deadline=[(i.split(" ")[1:4]) for i in deadline]
deadline 
```

```python
category=[] 

for url in allURLs:
    driver.get(url)
    driver.implicitly_wait(20)
    b = driver.find_elements_by_css_selector("span.ml1")[1].text 

    category.append(b)
    
print(category)
    

```

```python
launch=[]
for url in allURLs:
    driver.get(url)
    driver.implicitly_wait(10)
    link=driver.find_element_by_partial_link_text("Updates")
    link.click() 
    time.sleep(15)
    date = driver.find_element_by_xpath("//div[contains(@class,'timeline__divider--launched')]/div/div/time").text
    launch.append(date)

print(launch)
```

```python
df=pd.DataFrame() 
df['goal'] = goal 
df['category']= category
df['launch_date'] = launch
df['deadlin_date'] = deadline 
df
```

```python
df['goal'] = df.goal.apply(lambda x:x.replace('$',''))
df['goal'] = df.goal.apply(lambda x:x.replace(',',''))
df['goal'] = df.goal.apply(pd.to_numeric)
df
```

```python
df['launch_date'] = df.launch_date.apply(lambda x:x.replace(',',' '))
df['launch_date'] = pd.to_datetime(df.launch_date)
```

```python
df['launch_date']=df['launch_date'].dt.strftime('%Y%m%d')
```

```python
df
```

```python
deadline_date = []
for d in df['deadlin_date']:
    deadline_date.append(pd.to_datetime('-'.join(d)).strftime('%Y%m%d'))
```

```python
df['deadlin_date'] = pd.Series(deadline_date)
```
